**Ajuste el programa de regresiones lineales para que calcule la desviación estándar (sy), el error estándar de la estimación (sy/x) y el coeficiente de correlación (r) para el siguiente grupo de datos:**

x: {1, 2, 3, 4, 5, 6, 7}

y: {0.2, 0.5, 1.8, 3.4, 5.7, 9, 13.8}

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([0.2, 0.5, 1.8, 3.4, 5.7, 9, 13.8])

a, b = np.polyfit(x, y, 1)
y_pred = a * x + b

plt.scatter(x, y, color='blue', label='Puntos en X y en Y')
plt.plot(x, y_pred, color='black', label='Regresión lineal')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Regresión Lineal por Mínimos Cuadrados')
plt.grid(True, color='gray', linewidth=0.1)
plt.show()

print(f"Ecuación: y = {a:.4f}x + {b:.1f}")

#Hallar datos

y_Prom = np.mean(y)
st = np.sum((y-y_Prom)**2)
sy = np.sqrt(st/(6))
sr = np.sum((y-b-a*x)**2)
e = np.sqrt(sr/4)
c = np.sqrt((st-sr)/st)*100

print(f"La desviación estándar Sy = {sy:.6f}")
print(f"El error estándar de estimación Sy/x = {e:6f}")
print(f"El coeficiente de correlación r = {c:.1f} %")
```
