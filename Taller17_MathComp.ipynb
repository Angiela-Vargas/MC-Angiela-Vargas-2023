Cree un programa de regresiones lineales por mínimos cuadrados que permita aplicar el modelo exponencial. Aplique el modelo al siguiente grupo de datos:

x: {1, 2, 3, 4, 5, 6}

y: {2.2, 3, 4.5, 6, 8.5, 12}

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
import math

x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([2.2, 3, 4.5, 6, 8.5, 12])
aux = []

for i in range(6):
  n = math.log10(y[i])
  aux.append(n)

ylog = np.array(aux)

def grafico (Elem1, Elem2,title):

  a, b = np.polyfit(Elem1, Elem2, 1)
  y_pred = a * Elem1 + b

  plt.scatter(Elem1, Elem2, color='blue', label='Puntos en X y en Y')
  plt.plot(Elem1, y_pred, color='black', label='Regresión lineal')
  plt.xlabel('X')
  plt.ylabel('Y')
  plt.legend()
  plt.title(title)
  plt.grid(True, color='gray', linewidth=0.1)
  plt.show()
  warnings.simplefilter('ignore', np.RankWarning)

  print(f"Ecuación: y = {a:.4f}x + {b:.1f}")

inicial = grafico(x,y,"Regresión Lineal por Mínimos Cuadrados")
print()
resultado = grafico(x,ylog,"Regresión Lineal por Mínimos Cuadrados con Modelo Exponencial")
```
