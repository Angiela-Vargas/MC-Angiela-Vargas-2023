**Cree dos programas de regresiones lineales por mínimos cuadrados que apliquen los modelos de ecuación de potencias y razón de crecimiento. Aplique los 4 modelos vistos hasta ahora al siguiente listado de puntos y determine cuál se ajusta mejor desde una perspectiva visual:**

x: {1, 3, 5, 7, 9, 11, 13, 15}

y: {8.5, 13, 15, 16, 17, 17.5, 18, 19}

```python
from re import X
from numpy.ma.core import log10
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter('ignore', np.RankWarning)

def grafica (valor1, valor2,valor3,titulo,label):
  plt.scatter(valor1, valor2, color='blue', label='Puntos en X y en Y')
  plt.plot(valor1, valor3, color='black', label=label)
  plt.xlabel('X')
  plt.ylabel('Y')
  plt.legend()
  plt.title(titulo)
  plt.grid(True, color='gray', linewidth=0.1)
  plt.show()

#Código regresión lineal:

x = np.array([1, 3, 5, 7, 9, 11, 13, 15])
y = np.array([8.5, 13, 15, 16, 17, 17.5, 18, 19])

a, b = np.polyfit(x, y, 1)
y_pred = a * x + b


#Código regresión exponencial:

y_Log = np.array(log10(y))


#Código regresión potencial:

logaritmo_X = np.array(log10(x))
logaritmo_Y = np.array(log10(y))

n, m = np.polyfit(logaritmo_X, logaritmo_Y, 1)
alp = 10**m
y_Pot = alp * (x ** n)


#Código regresión por razón de crecimiento:

x_Prima = np.array(1/x)
y_Prima = np.array(1/y)

o, p = np.polyfit(x_Prima, y_Prima, 1)
k = 1/p
beta = o/p
y_Rcrecimiento = k*(x/(beta+x))


grafica_Inicial = grafica(x, y, y,"Representacion Gráfica de los Datos","Línea")
grafica_Lineal = grafica(x, y, y_pred,"Regresión Lineal por Mínimos Cuadrados","Regresión lineal")
print(f"Ecuación: y = {a:.4f}x + {b:.4f}")

grafica_Exponencial = grafica(x, y_Log, y_pred,"Regresión Exponencial", "Línea exponencial")

grafica_Potencias = grafica(x, y, y_Pot,"Regresión Ecuación de Potencias","Línea potencial")
ecuacion_Potencial = print(f"Ecuación: y = {alp:.4f}*x^{n:.4f}")

grafica_Rcrecimiento = grafica(x, y, y_Rcrecimiento, "Regresión Razón de crecimiento", "Línea de razón")
ecuacionPotencial = print(f"Ecuación: y = {k:.4f}*x/({beta:.4f}+x)")
```
