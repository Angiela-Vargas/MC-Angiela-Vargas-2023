Cree un programa que aplique regresiones lineales por mínimos cuadrados y utilícelo para ajustar una línea recta a:

X: 0, 1, 2, 3, 4, 5, 6, 7

Y: 7, 5, 6, 3, 4, 2.5, 6, 2, 0.5

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings

x = np.array([0, 1, 2, 3, 4, 5, 6, 7])
y = np.array([7, 5, 6, 3, 4, 2.5, 6, 2.5])

df = pd.read_csv('Taller16_MathComp.csv')
print(df)

a, b = np.polyfit(x, y, 1)
y_pred = a * x + b

plt.scatter(x, y, color='blue', label='Puntos en X y en Y')
plt.plot(x, y_pred, color='black', label='Regresión lineal')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Regresión Lineal por Mínimos Cuadrados')
plt.grid(True, color='gray', linewidth=0.1)
plt.show()
warnings.simplefilter('ignore', np.RankWarning)

print(f"Ecuación: y = {a:.4f}x + {b:.1f}")
```
